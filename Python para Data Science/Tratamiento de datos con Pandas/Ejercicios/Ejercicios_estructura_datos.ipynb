```python
# En primer lugar importamos pandas y numpy, la librería pathlib para recoger a la ruta del csv
import pandas as pd
import numpy as np 
from pathlib import Path 

ruta=Path("../datos/house.csv")

data_path=Path('precios.csv')

# Definimos el dataframe utilizando la función read_csv de Pandas
df_casas=pd.read_csv(filepath_or_buffer=ruta, sep=',')
```

```python
# Hallamos el número máximo y mínimo de habitaciones utilizando las funciones max() y min()
max_hab=max(df_casas['bedrooms'])
max_hab
```

```python
min_hab=min(df_casas['bedrooms'])
min_hab
```

```python
# Creamos una nueva columna que sea la relación de precio por planta

df_casas['price_floor']=round(df_casas['price']/df_casas['floors'],2)
# df_casas
```

# Ejercicio 4


- Convierte la columna "date" en un tipo fecha.

```python
# from datetime import date, datetime
```

```python
df_casas['date']=df_casas['date'].apply(pd.Timestamp)
df_casas['date']

```

- Genera tres nuevas columnas fruto de extraer el año, mes y día.

```python
df_casas['day']=df_casas['date'].dt.day
df_casas['month']=df_casas['date'].dt.month_name()
df_casas['year']=df_casas['date'].dt.year
df_casas.head()
```

- ¿Cual es el precio medio por mes?

```python
df_casas.set_index(df_casas['date'])
```

```python
df_casas.set_index(df_casas['date']).resample('1M')['price'].mean()
```

- Crea una nueva columna fruto de sumar 20 días a la columna fecha.

```python
td= pd.Timedelta('20 days')
df_casas['twenty_days_later']=df_casas['date']+td
df_casas
```
