## Ejercicios de Seaborn

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
data=sns.load_dataset('penguins')
data.head()
```

## Ejercicio 1

Genera un gráfico de regresión usando las variables 'bill_length_mm' y 'flipper_length_mm'. Implementa este gráfico separado por la variable 'sex'.


```python
# fif, ax= plt.subplots(figsize=(10,5))

sns.lmplot(data=data,x='bill_length_mm', y='flipper_length_mm', hue='sex')
```

Si implementamos el gráfico sin separar por sexo, podemos usar 'regplot'. 

```python
sns.regplot(data=data,x='bill_length_mm', y='flipper_length_mm')
```

## Ejercicio 2

¿Existe algún valor anómalo (outlier) en alguna variable?

```python
col_num=['bill_length_mm','bill_depth_mm', 'flipper_length_mm','body_mass_g']
```

```python
fig, ax=plt.subplots(ncols=4, figsize=(15,6))

i=0
for c in col_num:
    sns.boxplot(data[c], ax=ax[i])
    i+=1
```

#### No existe valor anómalo, ya que ningún valor se sale de los límites marcados
